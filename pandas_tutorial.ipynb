### Gli indici

In pandas ad ogni riga √® associato un indice, che pu√≤ essere numerico o di altro tipo. L'indice pu√≤ essere univoco o ripetuto. Se non viene specificato, l'indice √® numerico e parte da 0. Quando si combinano serie o dataframe, l'indice viene utilizzato per allineare i dati.

```python
s1 = pd.Series([1, 2, 3, 4, 5], index=[1, 4, 6, 8, 9])
s2 = pd.Series([10, 20, 30, 40, 50], index=[1, 2, 3, 4, 5])

s1 + s2
```

```python
# gli indici possono essere duplicati anche se generalmente non √® una buona pratica
s1 = pd.Series([1, 2, 3, 4, 5], index=[1, 1, 6, 8, 9]); s1
```

## Indexing

### iloc

la funzione `iloc` permette di selezionare righe e colonne utilizzando indici numerici.

```python
# seleziona la prima riga del dataframe
df.iloc[0, :]
```

```python
# seleziona le righe dalla seconda alla quarta
df.iloc[1:4, :]
```

```python
# seleziona la prima colonna delle righe dalla seconda alla quarta
df.iloc[1:4, 0]
```

### loc

la funzione `loc` permette di selezionare righe e colonne utilizzando come identificativi gli indici delle righe e i nomi delle colonne.

```python
# indexing con lable (loc)
df
```

```python
# dataframe
df.loc['2018-01-04':, ['numbers']]
```

```python
# series
df.loc['2018-01-04':, 'numbers']
```

### at e iat

```python
# accesso veloce a singoli elementi con at e iat
df.at['2018-01-04', 'numbers']
```

```python
df.iat[5, 1]
```

### indicizzazione booleana

```python
# maschere
df['numbers'] > 5
```

```python
# si pu√≤ fare anche con loc e iloc
len(df[df['numbers'] > 5])
```

```python
# WARNING!!!!!

# le parentesi sono necessarie perch√© gli operatori logici hanno una priorit√†
# maggiore rispetto a quelli di confronto

len(df[(df['numbers'] > 5) & (df['numbers'] < 8)])
```

```python
# WARNING

# vanno utilizzati gli operatori logici element-wise di Pandas

len(df[(df['numbers'] > 5) and (df['numbers'] < 8)])
```

## Groupby

```python
# https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html
```

```python
n = 1000
days = pd.date_range('2018-01-01', periods=n, freq='D')
names = [random.choice(['Angelo', 'Marco', 'Vittorio']) for _ in range(n)]
numbers = [random.randint(1, 10) for _ in range(n)]
ages = [random.randint(30, 80) for _ in range(n)]

df = pd.DataFrame(data=zip(names, numbers, ages), columns=['names', 'numbers', 'ages'], index=days)
```

```python
df.head()
```

```python
grouped = df.groupby('names')

for name, group in grouped:
    print(name)
    print(group)
```

```python
grouped.get_group('Angelo')
```

<img src="imgs\aggregazione.png" width=400>

```python
grouped.size()
```

```python
grouped.describe()
```

```python
grouped.std()
```

```python
## Aggregazione
df.groupby('names').agg(lambda x: sum(x))
```

```python
# Trasformazione
df.groupby(['names']).transform(lambda x: (x - x.mean()) / x.std())
```

```python
# Filtraggio
df.groupby(['names']).filter(lambda x: x['numbers'].sum() < 1940)
```
