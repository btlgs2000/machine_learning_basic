# Dataframe

Un DataFrame è la struttura dati principale di pandas, progettata per gestire dati tabulari. Alcune caratteristiche chiave:

- È composto da righe e colonne, dove ogni colonna può contenere dati di diverso tipo (numeri, testo, date, ecc.)
- Le colonne hanno un nome che le identifica
- Le righe sono identificate da un indice (numerico o di altro tipo)
- È bidimensionale: ha 2 assi (righe e colonne)

# Modalità di creazione

```python
n = 10
days = pd.date_range('2018-01-01', periods=n, freq='D')
names = [random.choice(['Angelo', 'Marco', 'Vittorio']) for _ in range(n)]
numbers = [random.randint(1, 10) for _ in range(n)]

df = pd.DataFrame(data=zip(names, numbers), columns=['names', 'numbers'], index=days)
```

La funzione `head()` permette di visualizzare le prime righe di un DataFrame.

La funzione `tail()` permette di visualizzare le ultime righe di un DataFrame.

```python
df.head()
```

Un Dataframe può essere creato da un dizionario di liste o array NumPy.



```python
# dataframe from dictionary
data = {
    'names': names,
    'numbers': numbers
}

pd.DataFrame(data).head()
```

Oppure da una lista di dizionari.

```python
data = [
    {'names': 'Angelo', 'numbers': 1},
    {'names': 'Marco', 'numbers': 2},
    {'names': 'Vittorio', 'numbers': 3}
]

pd.DataFrame(data).head()
```

Dataframe da array NumPy.

```python
data = np.random.randn(5, 3)
pd.DataFrame(data, columns=['A', 'B', 'C']).head()
```

# Gli indici

In pandas ad ogni riga è associato un indice, che può essere numerico o di altro tipo. L'indice può essere univoco o ripetuto. Se non viene specificato, l'indice è numerico e parte da 0. Quando si combinano serie o dataframe, l'indice viene utilizzato per allineare i dati.

```python
s1 = pd.Series([1, 2, 3, 4, 5], index=[1, 4, 6, 8, 9])
s2 = pd.Series([10, 20, 30, 40, 50], index=[1, 2, 3, 4, 5])

s1 + s2
```

Gli indici possono essere duplicati anche se generalmente non è una buona pratica

```python
s1 = pd.Series([1, 2, 3, 4, 5], index=[1, 1, 6, 8, 9]); s1
```

# Series

Una serie è una struttura dati unidimensionale, che può rappresentare una riga o una colonna di un dataframe.

Caratteristiche principali:

- È composta da un array NumPy e da un indice
- Ha un nome
- Ha un tipo di dati uniforme

```python
# creazione di una serie
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'], dtype='int64')
s
```

Una serie può rappresentare una colonna di un dataframe. In questo caso gli indici sono gli stessi del dataframe
e il nome della serie è il nome della colonna.

```python
df['LotArea']
```

Una serie che rappresenta una riga di un dataframe.
In questo caso gli indici sono i nomi delle colonne
e il nome della serie è l'indice della riga

```python
df.loc[1]
```

### Indicizzazione booleana

Se si applica un operatore di confronto a una serie o a un dataframe, viene restituita una serie o un dataframe di valori booleani.

```python
# maschere
df['LotArea'] > 10000
```

Una serie booleana può essere utilizzata per filtrare i dati di una serie o di un dataframe.

```python
df[df['LotArea'] > 10000]
```

Più serie booleane possono essere combinate con gli operatori logici `&` (and), `|` (or) e `~` (not).

Attenzione: questi operatori hanno alta priorità rispetto agli operatori di confronto, quindi è necessario utilizzare le parentesi per evitare errori.

```python
df[(df['LotArea'] > 10000) & (df['SalePrice'] > 200000)]
```

Gli operatori logici di python (`and`, `or`, `not`) non funzionano con le serie booleane di pandas.

```python
df[(df['LotArea'] > 10000) and (df['SalePrice'] > 200000)]
```

## Groupby

```python
# https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html
```

```python
n = 1000
days = pd.date_range('2018-01-01', periods=n, freq='D')
names = [random.choice(['Angelo', 'Marco', 'Vittorio']) for _ in range(n)]
numbers = [random.randint(1, 10) for _ in range(n)]
ages = [random.randint(30, 80) for _ in range(n)]

df = pd.DataFrame(data=zip(names, numbers, ages), columns=['names', 'numbers', 'ages'], index=days)
```

```python
df.head()
```

```python
grouped = df.groupby('names')

for name, group in grouped:
    print(name)
    print(group)
```

```python
grouped.get_group('Angelo')
```

<img src="imgs\aggregazione.png" width=400>

```python
grouped.size()
```

```python
grouped.describe()
```

```python
grouped.std()
```

```python
## Aggregazione
df.groupby('names').agg(lambda x: sum(x))
```

```python
# Trasformazione
df.groupby(['names']).transform(lambda x: (x - x.mean()) / x.std())
```

```python
# Filtraggio
df.groupby(['names']).filter(lambda x: x['numbers'].sum() < 1940)
```
