# Matplotlib

**Matplotlib** è una libreria Python per la creazione di grafici 2D. Ci sono due tipi di oggetti fondamentali: **Figure** e **Axes**. 

- **Figure** rappresenta l'intero contenitore del grafico, equivalente a una "pagina" o "finestra" su cui vengono disegnati i grafici. Può contenere uno o più sottografici.
- **Axes**, invece, è l'area specifica all'interno della figura dove vengono effettivamente tracciati i dati, con assi X e Y, etichette, titoli e altri elementi grafici.

## subplots

Il metodo `subplots` di `matplotlib.pyplot` crea una nuova figura e una griglia di sottografici (assi).

Creiamo una figura con un solo asse per plottare la funzione seno.


```python
fig, ax = plt.subplots(figsize=(6, 4))

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
ax.plot(x, y, label='Sine Wave')
ax.set_title('Sine Function')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')

# Set x-axis ticks to multiples of pi/2
ax.set_xticks(np.arange(0, 2 * np.pi + np.pi/2, np.pi/2))
ax.set_xticklabels([r'$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])

# Set minor ticks to multiples of pi/8
ax.set_xticks(np.arange(0, 2 * np.pi + np.pi/8, np.pi/8), minor=True)

# Show grid
ax.grid(True, 'both')

ax.legend()
plt.show()
```

# assi multipli

Creiamo una figura con due assi per plottare la funzione seno e la funzione coseno.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Plot Sine Wave
ax1.plot(x, y, label='Sine Wave')
ax1.set_title('Sine Function')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
ax1.set_xticks(np.arange(0, 2 * np.pi + np.pi/2, np.pi/2))
ax1.set_xticklabels([r'$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
ax1.set_xticks(np.arange(0, 2 * np.pi + np.pi/8, np.pi/8), minor=True)
ax1.grid(True, 'both')
ax1.legend()

# Plot Cosine Wave
y_cos = np.cos(x)
ax2.plot(x, y_cos, label='Cosine Wave', color='orange')
ax2.set_title('Cosine Function')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')
ax2.set_xticks(np.arange(0, 2 * np.pi + np.pi/2, np.pi/2))
ax2.set_xticklabels([r'$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
ax2.set_xticks(np.arange(0, 2 * np.pi + np.pi/8, np.pi/8), minor=True)
ax2.grid(True, 'both')
ax2.legend()

plt.show()
```

```python
# Creiamo dei dati di esempio
np.random.seed(42)
x = np.linspace(0, 10, 50)
y = np.random.randn(50)
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
matrix = np.random.rand(20, 20)

# Creiamo una figura con 3x2 subplots
fig, axs = plt.subplots(3, 2, figsize=(15, 18))

# Titolo della figura
fig.suptitle('Esempi di grafici', fontsize=20)

# 1. Line plot
axs[0, 0].plot(x, np.sin(x), 'b-', label='sin(x)')
axs[0, 0].plot(x, np.cos(x), 'r--', label='cos(x)')
axs[0, 0].set_title('Line Plot')
axs[0, 0].legend()
axs[0, 0].grid(True)

# 2. Scatter plot
axs[0, 1].scatter(x, y, c=y, cmap='viridis', s=100)
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].grid(True)

# 3. Bar plot
axs[1, 0].bar(categories, values, color='skyblue')
axs[1, 0].set_title('Bar Plot')

# 4. Histogram
axs[1, 1].hist(y, bins=15, color='green', alpha=0.7)
axs[1, 1].set_title('Histogram')

# 5. Heatmap
im = axs[2, 0].imshow(matrix, cmap='hot')
axs[2, 0].set_title('Heatmap')
plt.colorbar(im, ax=axs[2, 0])

# 6. Pie chart
axs[2, 1].pie(values, labels=categories, autopct='%1.1f%%')
axs[2, 1].set_title('Pie Chart')

plt.tight_layout()
```

# Seaborn

**Seaborn** è una libreria di alto livello basata su Matplotlib progettata per rendere la creazione di grafici statistici più semplice e visivamente accattivante. Fornisce API di alto livello per realizzare grafici complessi con poche righe di codice, integrandosi con i DataFrame di **Pandas** per gestire i dati in modo più naturale. 

### Vantaggi rispetto a Matplotlib:
1. **Grafici Predefiniti e Stilizzati**: Seaborn offre uno stile preconfigurato e accattivante, eliminando la necessità di personalizzazioni manuali.
2. **Gestione Automatica dei Dati**: È possibile creare grafici direttamente dai DataFrame, grazie al riconoscimento automatico delle variabili.
3. **Visualizzazioni Statistiche**: Seaborn include funzioni per analisi statistiche, come **distribuzioni** (`sns.histplot`), **relazioni** (`sns.relplot`), e **matrici di correlazione** (`sns.heatmap`), che richiederebbero configurazioni manuali in Matplotlib.

Seaborn ha tre categorie di grafici:

1. **Relational Plots**: Mostrano la relazione tra due variabili, come scatter plot, line plot, ecc.
2. **Categorical Plots**: Mostrano una variabile categorica.
3. **Distribution Plots**: Mostrano la distribuzione di una variabile.

Per ciascuna categoria, seaborn offre una fuzione principale di livello superiore (figura) e diverse funzioni di livello inferiore (assi) per creare grafici specifici.

# Plotly

**Plotly** è una libreria Python per la creazione di grafici interattivi, ideale per esplorare i dati. I grafici generati permettono di effettuare **zoom**, **pan**, e **ispezionare i dati** tramite tooltip che mostrano informazioni dettagliate direttamente sui punti selezionati. 

Vantaggi:
1. **Interattività**: I grafici Plotly sono interattivi e permettono di esplorare i dati in modo dinamico.

Svantaggi:
1. **Difficoltà nella gestione di grandi dataset**: Plotly può diventare lento con dataset molto grandi, a causa dell'overhead necessario per la gestione dell'interattività.
